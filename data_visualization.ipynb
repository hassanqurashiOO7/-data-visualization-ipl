```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
df = pd.read_csv('IPL.csv')
df
```

#Whic team won most matches

```python
df.head(5)
```

```python
won_most_matches = df['match_winner'].value_counts()

sns.barplot(
    y=won_most_matches.index,
    x=won_most_matches.values,
    hue=won_most_matches.index,  
    palette='viridis',
    legend=False               
)

plt.title('Most Won Matches')
plt.show()
```

toss decision

```python
sns.countplot(y=df['toss_winner'],
             hue=df['toss_winner'],
             palette ='viridis',
             legend=False
             )
plt.title('Most Won Toss')
plt.show()
```

toss winner vs match winner

```python
bt=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(bt*100)/df.shape[0]
percentage


```
